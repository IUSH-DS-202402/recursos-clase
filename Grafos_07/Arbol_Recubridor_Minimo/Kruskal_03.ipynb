# Algoritmo de Kruskal



Primero creamos la lista de vértices, la lista de lados y la ordenamos.

```python
listaVertices = ["A", "B", "C", "D", "E"]
listaLados = [[1, "A", "B"], 
              [7, "A", "C"], 
              [10, "A", "D"], 
              [5, "A", "E"], 
              [3, "B", "C"], 
              [4, "C", "D"], 
              [2, "D", "E"] ]
```

```python
from functools import cmp_to_key
from collections import deque 

def cmpr(item1, item2): 
    return item1[0] - item2[0]
```

```python
from DisjSet import DisjSet
def kruskal(listaVertices,listaLados):
    # Se ordenan los lados ascendentemente por peso
    listaLadosOrdenados = sorted(listaLados, key=cmp_to_key(cmpr))
    # Se crea una cola con los datos ordenados
    Ed = deque(listaLadosOrdenados)
    # Se crea el conjunto disyunto de vertices 
    V = DisjSet(listaVertices) 
    st = [] # Se inicia el árbol sin datos
    i = 0
    # Mientras el árbol no esté completo
    while i < len(listaVertices)-1 and len(Ed) > 0:
        # Se desencola el lado con el menor peso de la cola.
        lado = Ed.popleft()
        # Si no se forma ciclo al agregar este lado al árbol.
        if V.find_kw(lado[1]) != V.find_kw(lado[2]):
            st.append(lado) # Se agrega el lado al spanning tree
            # Se unen los conjuntos a los que pertenecen ambos vertices
            V.Union(lado[1],lado[2]) 
            i += 1
    #Fin del while
    return st
```

```python
x = kruskal(listaVertices,listaLados)
print(x)
```
